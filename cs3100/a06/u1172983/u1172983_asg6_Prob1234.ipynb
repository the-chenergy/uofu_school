* First of all, the function `mk_plus_nfa` works because it follows the definition of NFA union: `Q`, `Sigma`, `q0`, and `F` are all the unions of the `Q`s, `Sigma`s, `q0`s, and `F`s, respectively, from the inputs, and it accumulates the `Delta`s.

* Then, the function `mk_cat_nfa` works because it does exactly what it takes to concatenate two NFAs: link all the final states from one NFA to the other's initial states going upon epsilon, while making those connected states non-final and non-initial.

* Then, the function `mk_star_nfa` works because it does exactly what it takes to "star" the contents of an NFA: create a new initial-final state, link all the original final states to that IF state upon epsilon, link this IF state to all the original initial states upon epsilon, and unflag the original initial/final states.

* Then, the function `mk_eps_nfa` works because an NFA which accepts and only accepts epsilon is one with one IF state.

* Finally, the function `mk_symbol_nfa` works because it does what it's meant to do: it simply returns an NFA that accepts the next symbol in the input sequence.
