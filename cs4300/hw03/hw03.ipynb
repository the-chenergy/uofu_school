## Part 2.4

According to the Law of Total Probability,

$$
\begin{aligned}
  \text{P}(\text{Weather = sun}) &= \text{P}(\text{Weather = sun, Temperature = cold})\\
  &\ \ \ \ + \text{P}(\text{Weather = sun, Temperature = warm})\\
  &= 0.3 + 0.2\\
  &= \boxed{0.5}
\end{aligned}
$$

## Part 2.5

$$
\begin{aligned}
  \text{P}(\text{Weather = snow}) &= \text{P}(\text{Weather = snow, Temperature = cold})\\
  &\ \ \ \ + \text{P}(\text{Weather = snow, Temperature = warm})\\
  &= 0.1 + 0.4\\
  &= 0.5\\
  \text{P}(\text{Weather = cold}) &= \text{P}(\text{Weather = snow, Temperature = cold})\\
  &\ \ \ \ + \text{P}(\text{Weather = sun, Temperature = cold})\\
  &= 0.1 + 0.3\\
  &= 0.4
\end{aligned}
$$

According to the Inclusion-Exclusion Principle,

$$
\begin{aligned}
  &\text{P}(\text{Weather = snow} \cup \text{Temperature = cold})\\
  &= \text{P}(\text{Weather = snow}) + \text{P}(\text{Temperature = cold}) - \text{P}(\text{Weather = snow}, \text{Temperature = cold})\\
  &= 0.5 + 0.4 - 0.1\\
  &= \boxed{0.8}
\end{aligned}
$$

# Problem 4

First, let's calculate the probability/proportion of time that the dog is barking using the Law of Total Probability:

$$
\begin{aligned}
  \text{P}(+b) &= \text{P}(+b,m) + \text{P}(+b,q) + \text{P}(+b,n)\\
  &= \text{P}(+b\mid m)\ \text{P}(m) + \text{P}(+b\mid q)\ \text{P}(q) + \text{P}(+b\mid n)\ \text{P}(n)\\
  &= 0.8 \times 0.1 + 0.9 \times 0.01 + 0.4 \times 0.89\\
  &= 0.445
\end{aligned}
$$

Therefore, the dog barks 44.5% of the time or in other words, the dog has a 44.5% chance of barking given any situation.

Now, let's use Bayes' Theorem to find the likelihood of events happening given the evidence of the dog barking:

$$
\begin{aligned}
  \text{P}(m\mid +b) &= \frac{\text{P}(+b\mid m)\ \text{P}(m)}{\text{P}(+b)} = \frac{0.8 \times 0.1}{0.445} \approx 18\%\\
  \text{P}(q\mid +b) &= \frac{\text{P}(+b\mid q)\ \text{P}(q)}{\text{P}(+b)} = \frac{0.9 \times 0.01}{0.445} \approx 2\%\\
  \text{P}(n\mid +b) &= \frac{\text{P}(+b\mid n)\ \text{P}(n)}{\text{P}(+b)} = \frac{0.8 \times 0.89}{0.445} \approx 80\%
\end{aligned}
$$

Since 80% of the time there's really nothing happening when the dog is barking, you should just ignore the dog. Though, as 80% isn't very confident, you should also try getting more evidence to help narrow down the situation.
