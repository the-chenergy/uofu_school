# Problem 1: Inference

## Part 1.1

$$\P(-y\mid +x) = 1 - \P(+y\mid +x) = 1 - .8 = \boxed{.2}$$

## Part 1.2

$$\P(-z\mid -x) = 1 - \P(+z\mid -x) = 1 - .4 = \boxed{.6}$$

## Part 1.3

$$\P(-y\mid +x, +z) = \P(-y\mid +x) = \boxed{.2}$$

## Part 1.4

$$
\begin{aligned}
  \P(+y, -z, -x) &= \P(+y, -z\mid -x)\ \P(-x)\\
                 &= \P(+y\mid -x)\ \P(-z\mid -x)\ \P(-x)\\
                 &= .3 \times .6 \times .7\\
                 &= \boxed{.126}
\end{aligned}
$$

## Part 1.5

$$
\begin{aligned}
  \P(-y) &= 1 - \P(+y)\\
         &= 1 - (\P(+y, +x) + \P(+y, -x))\\
         &= 1 - (\P(+y\mid +x)\ \P(+x) + \P(+y\mid -x)\ \P(-x))\\
         &= 1 - (.8 \times .3 + .3 \times .7)\\
         &= \boxed{.55}
\end{aligned}
$$

## Part 1.6

$$
\begin{aligned}
  \P(+y, +z) &= \P(+y, +z, +x) + \P(+y, +z, -x)\\
             &= \P(+y\mid +z, +x)\ \P(+z, +x) + \P(+y\mid +z, -x)\ \P(+z, -x)\\
             &= \P(+y\mid +x)\ \P(+z\mid +x)\ \P(+x) + \P(+y\mid -x)\ \P(+z\mid -x)\ \P(-x)\\
             &= .8 \times .8 \times .3 + .3 \times .4 \times .7\\
             &= .276\\
  \\
  \P(+z) &= \P(+z, +x) + \P(+z, -x)\\
         &= \P(+z\mid +x)\ \P(+x) + \P(+z\mid -x)\ \P(-x)\\
         &= .8 \times .3 + .4 \times .7\\
         &= .52\\
  \\
  \P(+y\mid +z) &= \frac{\P(+y, +z)}{\P(+z)}\\
                &= \frac{.276}{.52}\\
                &\approx \boxed{.5308}
\end{aligned}
$$

## Part 1.7

$$
\begin{aligned}
  \P(-y\mid +z) &= 1 - \P(+y\mid +z)\\
                &\approx 1 - .5308\\
                &\approx \boxed{.4692}
\end{aligned}
$$

$$

  R & T & \P(R, T)\\
  \hline

$$

  \hline
\end{array}

$$
\begin{array}{|cc|r|}

# Problem 5: Sampling

## Part 5.1

$$\text{w}(+s, +w, +c, +r) = \P(+c)\ \P(+r\mid +c) = .5 \times .8 = \boxed{.4}$$

## Part 5.2

$$\text{w}(+s, +w, -c, +r) = \P(-c)\ \P(+r\mid -c) = .5 \times .2 = \boxed{.1}$$

## Part 5.3

$$\text{w}(+s, +w, +c, -r) = \P(+c)\ \P(-r\mid +c) = .5 \times .2 = \boxed{.1}$$

## Part 5.4

$$\text{w}(+s, +w, -c, -r) = \P(-c)\ \P(-r\mid -c) = .5 \times .8 = \boxed{.4}$$

## Part 5.5

The total value of the samples is
$$.4 + .1 + .1 + .4 + .4 = 1.4$$

Therefore,
$$
\begin{aligned}
  \P(+s, +w, +c, +r) &= \frac{1 \times .4}{1.4} \approx .2857\\
  \P(+s, +w, +c, -r) &= \frac{0 \times .1}{1.4} = 0\\
  \P(+s, +w, +c) &= \P(+s, +w, +c, +r) + \P(+s, +w, +c, -r) \approx .2857 + 0 \approx \boxed{.2857}
\end{aligned}
$$

## Part 5.6

$$
\begin{aligned}
  \P(+s, +w, -c, +r) &= \frac{2 \times .1}{1.4} \approx .1429\\
  \P(+s, +w, -c, -r) &= \frac{2 \times .4}{1.4} \approx .5714\\
  \P(+s, +w, -c) &= \P(+s, +w, -c, +r) + \P(+s, +w, -c, -r) \approx .1429 + .5714 \approx \boxed{.7143}
\end{aligned}
$$

# Problem 6: Decision Networks

## Part 6.1

$$
\begin{aligned}
  \text{EU(take)} &= \text{U(take, sun)}\ \P(\text{sun}) + \text{U(take, rain)}\ \P(\text{rain})\\
                  &= 20 \times .6 + 70 \times .4\\
                  &= \boxed{40}
\end{aligned}
$$

## Part 6.2

$$
\begin{aligned}
  \text{EU(leave)} &= \text{U(leave, sun)}\ \P(\text{sun}) + \text{U(leave, rain)}\ \P(\text{rain})\\
                  &= 80 \times .6 + 0 \times .4\\
                  &= \boxed{48}
\end{aligned}
$$

## Part 6.3

Since EU(leave) $>$ EU(take), it is better to leave the umbrella behind.

## Part 6.4

$$
\begin{array}{|c|r|}
  \hline
  F & \P(F)\\
  \hline
  \text{good} & = \P(\text{good | sun})\ \P(\text{sun}) + \P(\text{good | rain})\ \P(\text{rain}) = .6 \times .7 + .4 \times .2 = .5\\
  \text{bad} & = \P(\text{bad | sun})\ \P(\text{sun}) + \P(\text{bad | rain})\ \P(\text{rain}) = .6 \times .3 + .4 \times .8 = .5\\
  \hline
\end{array}
$$
$$
\begin{array}{|cc|r|}
  \hline
  W & F & \P(W\mid F)\\
  \hline
  \text{sun} & \text{good} & = \frac{\P(\text{sun})\ \P(\text{good | sun})}{\P(\text{good})} = \frac{.6 \times .7}{.5} = .84\\
  \text{sun} & \text{bad} & = \frac{\P(\text{sun})\ \P(\text{bad | sun})}{\P(\text{bad})} = \frac{.6 \times .3}{.5} = .36\\
  \text{rain} & \text{good} & = \frac{\P(\text{rain})\ \P(\text{good | rain})}{\P(\text{good})} = \frac{.4 \times .2}{.5} = .16\\
  \text{rain} & \text{bad} & = \frac{\P(\text{rain})\ \P(\text{bad | rain})}{\P(\text{bad})} = \frac{.4 \times .8}{.5} = .64\\
  \hline
\end{array}
$$
$$
\begin{aligned}
  \text{EU(take | good)} &= \text{U(take, sun)}\ \P(\text{sun | good}) + \text{U(take, rain)}\ \P(\text{rain | good})\\
                  &= 20 \times .84 + 70 \times .16\\
                  &= \boxed{28}
\end{aligned}
$$

## Part 6.5

$$
\begin{aligned}
  \text{EU(leave | good)} &= \text{U(leave, sun)}\ \P(\text{sun | good}) + \text{U(leave, rain)}\ \P(\text{rain | good})\\
                  &= 80 \times .84 + 0 \times .16\\
                  &= \boxed{67.2}
\end{aligned}
$$

## Part 6.6

Since EU(leave | good) $>$ EU(take | good), it is better to leave the umbrella behind.
